```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import src.evalModel_fx1 as emu_fx1
import src.evalModel_fxLeq1 as emu_fxLeq1


kk = np.logspace(np.log10(0.01), np.log10(10), 1000)


cosmo_par = {
    
'Omega_b':0.0486,
'Omega_cdm':0.2617,
'm_ncdm':0,
'n_s':0.9660,
'h': 0.6755,
'w0_fld':-0.9,
'log_cs2': -7 ,
'A_s':2.21e-9,
    
}

z = [0,0.2,0.4]


kk, mu_fx1 = emu_fx1.get_mu(cosmo_par,z)
kk, mu_fxLeq1 = emu_fxLeq1.get_mu(cosmo_par,z)
```

```python
plt.rc("text", usetex=True)
font = {"family": "normal", "size": 15}

colors = ['blue','orange', 'green']

for i in range(len(z)):
    plt.loglog(kk,mu_fx1[i],label=r"$\mu_\mathrm{f_x=1}~z=$"+str(z[i]),ls = '--', c=colors[i])
    plt.loglog(kk,mu_fxLeq1[i],label=r"$\mu_\mathrm{f_x\leq1}~z=$"+str(z[i]), c=colors[i], linestyle = '-')


plt.xlabel(r'$k\ [h/\rm{Mpc}]$', fontsize = 15); plt.ylabel(r'$\mu(k,z)$', fontsize = 15)
plt.legend(fontsize = 14)
```
